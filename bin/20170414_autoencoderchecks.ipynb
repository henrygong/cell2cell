```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

recon = np.load("cgmcodae_reconstructions.npy")
```

```python
org = np.load("../../data/20170413/cgmco.npy")
```

```python
from sklearn.decomposition import PCA

all_data = np.vstack([recon, org])
pca = PCA(n_components=2)
pca_tx = pca.fit_transform(all_data)

plt.scatter(pca_tx[:, 0], pca_tx[:, 1])
```

```python
colors = ['green' if i < recon.shape[0] else 'blue' for i in range(all_data.shape[0])]
plt.scatter(pca_tx[:, 0], pca_tx[:, 1], c=colors, linewidths=0., alpha=0.5)
```
